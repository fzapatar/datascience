```python
#Pregunta 1

import yfinance as yf

# Descargar datos bursátiles de Tesla
tesla_data = yf.download("TSLA", period="max")

# Restablecer el índice
tesla_data.reset_index(inplace=True)

# Visualizar las primeras 5 filas
print(tesla_data.head())
```

```python
#Pregunta 2
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, 'html.parser')

# Extraer tabla de ingresos
table = soup.find_all("table")[1]  # La tabla objetivo
rows = table.find_all("tr")
tesla_revenue = pd.DataFrame([
    [td.text.strip() for td in row.find_all("td")] for row in rows[1:]
], columns=["Date", "Revenue"])

# Mostrar las últimas 5 filas
print(tesla_revenue.tail())

```

```python
#Pregunta 3
# Descargar datos bursátiles de GameStop
gme_data = yf.download("GME", period="max")

# Restablecer el índice
gme_data.reset_index(inplace=True)

# Visualizar las primeras 5 filas
print(gme_data.head())
```

```python
#Pregunta 4
url_gme = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
html_data_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(html_data_gme, 'html.parser')

# Extraer tabla de ingresos
table_gme = soup_gme.find_all("table")[1]
rows_gme = table_gme.find_all("tr")
gme_revenue = pd.DataFrame([
    [td.text.strip() for td in row.find_all("td")] for row in rows_gme[1:]
], columns=["Date", "Revenue"])

# Mostrar las últimas 5 filas
print(gme_revenue.tail())
```

```python
#Pregunta 5 y 6
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Date'], data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.grid(True)
    plt.show()

# Graficar Tesla
make_graph(tesla_data, "Tesla Stock Prices Over Time")

# Graficar GameStop
make_graph(gme_data, "GameStop Stock Prices Over Time")
```
